**COMPARISON OF NET EMISSIONS IN LAST 20 YEARS**

```python
china=Econ[Econ["Country"]=="China"]["Net Emissions"].iloc[0]
us=Econ[Econ["Country"]=="United States"]["Net Emissions"].iloc[0]
ind=Econ[Econ["Country"]=="India"]["Net Emissions"].iloc[0]
chinanew=Econ[Econ["Country"]=="China"]["2017"].iloc[0]
usnew=Econ[Econ["Country"]=="United States"]["2017"].iloc[0]
indnew=Econ[Econ["Country"]=="India"]["2017"].iloc[0]
```

```python
def percent(x,y):
    if x>y:
        a=(((x-y)*100)/y)
        return(int(a))
    elif x<y:
        b=((y-x)*100)/x
        return(int(b))
        
    
```

```python
percent(china,ind)
```

**Comparisons With World** 

```python
world=data[data["Country"]=="World"]["Net Emissions"].iloc[0]
worldnew=data[data["Country"]=="World"]["2017"].iloc[0]
```

```python
#Writing Functions That Give Percentage Share Of Countries In Global Emissions
def percentofworld(a):
    return(int((a*100)/world))
def percentofworldnew(a):
    return(int((a*100)/worldnew))
```

```python
percentofworld(us)
```

The United States Share Of Global Emissions Of Harmful Gases Is 27% In Last 20 Years

```python
percentofworldnew(usnew)
```

The United States Share Of Global Emissions Of Harmful Gases Shrunk To 25% In 2017

```python
percentofworld(china)
```

The Chinese Share Of Global Emissions Of Harmful Gases Was 9% In Last 20 Years

```python
percentofworldnew(chinanew)
```

The Chinese Share Of Global Emissions Of Harmful Gases Rose To 12% In 2017

```python
percentofworld(ind)
```

```python
percentofworldnew(indnew)
```

```python
restofworld=world-china-us-ind
```

```python
my_data = lis1
my_labels = 'India','China','USA','Rest Of World'
my_explode = (0, 0, 0,0.1)
plt.pie(my_data,labels=my_labels,autopct='%1.1f%%',startangle=30, shadow = True, explode=my_explode,colors=['lightblue','lightgrey','lightsteelblue','gold'])
plt.title('GLOBAL SHARE OF EMISSIONS IN LAST 20 YEARS')
plt.axis('equal')
plt.show()
```
