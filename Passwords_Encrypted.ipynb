```python
import bcrypt
import pandas as pd
```

Initialize a dictionary to store {mail: hashed_password} pairs

# Generate a salt and hash the password securely\
Generating a salt and securely hashing passwords is a standard practice for protecting user data by preventing plain text password storage, making it difficult for attackers to retrieve passwords, and ensuring uniqueness in password hashes.

```python
hashed_passwords = {}

def add_password(mail, password):
    # Generate a salt and hash the password securely
    salt = bcrypt.gensalt()
    hashed_password = bcrypt.hashpw(password.encode('utf-8'), salt)  # Declare hashed_password here

    # Store the hashed password in the dictionary
    hashed_passwords[mail] = hashed_password
```

Crete the dictionary

```python
mypasswords = {
    "gmail": "mygmailpassword",
    "trebas": "mytrebaspassword",
    "facebook": "myfacebookpassword"
}
```

Hash the passwords and store them

```python
for mail, password in mypasswords.items():
    add_password(mail, password)
```

```python
for mail, hashed_password in hashed_passwords.items():
    print(f"Mail: {mail}, Hashed Password: {hashed_password}")
```
